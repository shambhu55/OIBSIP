## Exploratory Data Analysis (EDA) on Retail Sales Data

In this project, I worked with a dataset containing retail sales information. The main objective is to conduct exploratory data analysis (EDA) to discover patterns, trends, and insights that can help the retail business make informed decisions. By thoroughly analyzing the data, we uncovered valuable insights that contribute to shaping business strategies and improving customer engagement.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Loading and Cleaning: Load the retail sales dataset.

```python
df = pd.read_csv(r"C:\Users\HP\Downloads\retail_sales_dataset.csv")
```

```python
# Calculate mean, median, and standard deviation for all numeric columns
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')     
```

```python
# Data Cleaning: Converting 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

### Descriptive Statistics: Calculate basic statistics (mean, median, mode, standard deviation).

```python
mean_values = numeric_df.mean()  # Mean
print("Mean:\n", mean_values)
```

```python
median_values = numeric_df.median()  # Median
print("\nMedian:\n", median_values)
```

```python
std_values = numeric_df.std()  # Standard deviation
print("\nStandard Deviation:\n", std_values)
```

```python
# Calculate mode (mode returns the most frequent value(s)) for numeric columns
mode_values = numeric_df.mode().iloc[0]  # Mode
print("\nMode:\n", mode_values)
```

### Time Series Analysis: Analyze sales trends over time using time series techniques.

```python

sns.relplot(df['Total Amount'],kind='scatter')
```

### Customer and Product Analysis: Analyze customer demographics and purchasing behavior.

```python
# Customer Demographics Analysis
customer_group = df.groupby('Customer ID').agg({
    'Gender': 'first',
    'Age': 'first',
    'Total Amount': 'sum',
    'Quantity': 'sum'
}).reset_index()
```


### Visualization: Present insights through bar charts, line plots, and heatmaps.

```python
# 1. Analyzing purchasing behavior by Gender
gender_sales = df.groupby('Gender')['Total Amount'].sum()

# Visualization: Bar chart for Total Sales by Gender
plt.figure(figsize=(6, 6))
sns.barplot(x=gender_sales.index, y=gender_sales.values, palette='Set2')
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# 2. Analyzing Age distribution and purchasing behavior
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Analyzing spending patterns by age groups
df['Age Group'] = pd.cut(df['Age'], bins=[18, 30, 40, 50, 60], labels=['18-30', '31-40', '41-50', '51-60'])
age_group_sales = df.groupby('Age Group')['Total Amount'].sum()

# Visualization: Line plot for Sales by Age Group
plt.figure(figsize=(8, 6))
sns.lineplot(x=age_group_sales.index, y=age_group_sales.values, marker='o', color='purple')
plt.title('Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# Product Analysis: Total sales by Product Category
product_sales = df.groupby('Product Category')['Total Amount'].sum()

# Visualization: Bar chart for Sales by Product Category
plt.figure(figsize=(6, 6))
sns.barplot(x=product_sales.index, y=product_sales.values, palette='Set1')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Amount Spent')
plt.xticks(rotation=45)
plt.show()
```
